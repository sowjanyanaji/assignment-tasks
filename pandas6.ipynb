# Missing values

```python
df.isna().sum()
```

```python
df['Age'].fillna(df['Age'].median(),inplace = True)
df['Gender'].fillna(df['Gender'].mode()[0],inplace = True)
df['Category'].fillna(df['Category'].mode()[0],inplace = True)
df['Item Purchased'].fillna(df['Item Purchased'].mode()[0],inplace = True)
df['Location'].fillna(method = 'bfill',inplace = True)
df['Color'].fillna(df['Color'].mode()[0],inplace = True)
df['Season'].fillna(df['Season'].mode()[0],inplace = True)
df['Review Rating'].fillna(df['Review Rating'].median(),inplace = True)
df['Subscription Status'].fillna(method = 'bfill',inplace = True)
df['Discount Applied'].fillna(method = 'ffill',inplace = True)
df['Promo Code Used'].fillna(df['Promo Code Used'].mode()[0],inplace = True)
df['Payment Method'].fillna(method = 'ffill',inplace = True)
df['Frequency of Purchases'].fillna(method = 'ffill',inplace = True)
```

# Scaling

```python
#if mean and median are closed values===then normal distribution,else standardization
```

```python
df["Age"].mean(),df["Age"].median()

```

```python
def normalization(x):
    n = (x-df["Age"].min())/(df["Age"].max()-df["Age"].min())
    return n
```

```python
df["Age"].apply(normalization)
```
