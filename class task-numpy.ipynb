# sepal width

```python
swid= flower[1:,1]
swid
```

```python
np.nanmedian(swid) #median
```

```python
swid = np.where(np.isnan(swid),3,swid)  #missing values
swid
```

```python
outlier_swid = len(swid[swid>6]) #no.of outliers in sepal width
outlier_swid
```

```python
final_swid = np.where(swid>6,4,swid)
final_swid
```

**sorting sepal width in ascending order**

```python
np.sort(final_swid)
```

```python
np.mean(final_swid)  #mean of sepal width
```

```python
np.median(final_swid) #median of sepal width
```

```python
np.max(final_swid) #largest width
```

```python
np.min(final_swid) #shortest sepal width
```

```python
range1 = np.max(final_swid)-np.min(final_swid)
range1
```

```python
np.var(final_swid)
```

```python
np.std(final_swid)
```

```python
np.savetxt("cleaned_sepalwidth",final_swid)
```

```python
np.loadtxt("cleaned_sepalwidth")
```
