```python
#outliers
```

we can check outliers through statiscal method for other than objet data type,box plot through visualisation

```python
df=  df1.select_dtypes(exclude="object")
for i in df.columns:
    Q1 = df[i].quantile(0.25)
    Q3 = df[i].quantile(0.75)

    IQR = Q3-Q1
    lower = Q1 - (1.5*IQR)
    upper = Q3 + (1.5*IQR)
    df[(df[i]<lower)] = lower
    df[(df[i]>upper)]= upper
    print(f"  {i} Lower_limit {lower} ")
    print(f"  {i} Upper_limit {upper} ")
    print()
```

```python
df['Sales'].plot(kind='box')
```
