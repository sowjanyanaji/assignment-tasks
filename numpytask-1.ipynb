**IQR**

```python
q1=np.percentile(array,25)
q1
```

```python
q2=np.percentile(array,50)
q2
```

```python
q3=np.percentile(array,75)
q3
```

```python
iqr=q3-q1
iqr
```

```python
lower=q1-(1.5*iqr)
lower
```

```python
upper=q3+(1.5*iqr)
upper
```
